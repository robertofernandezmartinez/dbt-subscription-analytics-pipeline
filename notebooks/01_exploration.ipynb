### `productprice` - Is the 0.01 price pattern concentrated in a specific product category?

If it's random noise it should be spread evenly. 

If it's systematic it's a business artifact — bundled services, migration placeholders, internal line items.

```python
products['product_id'] = pd.to_numeric(products['product_id'], errors='coerce')
low_price = orders[orders['productprice'] == 0.01]
low_price.merge(products[['product_id', 'product_cat_desc,,,,']], on='product_id', how='left')['product_cat_desc,,,,'].value_counts()
```

Pattern is present across all categories but heavily concentrated in Applications (63% of all 0.01 rows) — consistent with app activations or included add-ons where the "price" is a system placeholder, not a real charge.

#### `productprice` Action

Flag as `is_revenue_valid = FALSE`, do NOT drop — these are real order records (but generate no meaningful recurring revenue)

### `productprice` - Do the 0.01 price rows and the quantity = 0 rows overlap?

If they're the same population, they represent one single class of non-revenue records, not two separate problems.

```python
overlap = ((orders['productprice'] == 0.01) & (orders['quantity'] == 0)).sum()
total_low = (orders['productprice'] == 0.01).sum()
total_zero_qty = (orders['quantity'] == 0).sum()
print(f"Both conditions: {overlap:,}")
print(f"price = 0.01 only: {total_low:,}")
print(f"quantity = 0 only: {total_zero_qty:,}")
```

- 49,201 rows have both `productprice = 0.01` AND `quantity = 0` — these are the same population

But they're not identical:

- 20,766 rows have `price = 0.01` with `quantity > 0`
- and 29,427 have `quantity = 0` with a real price

They overlap heavily but are not the same thing — both conditions are needed independently in the `is_revenue_valid` flag, not just one of them

#### `productprice` Action

Exclude `quantity = 0` orders from MRR — `productprice × quantity = 0` regardless of price, no recurring revenue generated. 

Implemented as `quantity = 0` condition in `is_revenue_valid` flag in staging.

- No action needed for MRR — the productprice / duration_months formula handles all values correctly as long as duration > 0 (already guarded in staging)

Flag to business: confirm whether 120-month and non-standard durations (25, 55, 70, 72, 84) are valid contract terms or data entry errors — no blocking issue for the model but worth validating

### Subscriptions - Summary

**`status`**
- Only 48,173 non-null out of 170,245 — 72% of rows have no status value
- **The most critical finding in this table**: NULL doesn't mean unknown, it means "nothing happened yet" — **need to derive active/churned status from `expire_date` instead**

**`subscription_id`**
- Clean PK
- In orders it was float64 because of the ~25k nulls — confirms those orders exist without a matching subscription

**`creation_date` / `start_date` / `expire_date`**
- All object (string) — **same casting issue as orders, need explicit date conversion before any churn/retention calculation**
  

#### Dates:

**`creation_date` / `start_date`**
- Data goes back to 1996-1998
- Both columns move together (means and percentiles almost identical) — creation and start are essentially the same event for most subscriptions

**`expire_date`**
- Min = 2016 - some subscriptions expired years ago — historical churn already in the data
- Pct 25% = 2022 — a quarter of subscriptions expired before 2022, already churned
- Median = 2024-07-25 — more than half of all subscriptions have already expired as of today (July 2026)
- Pct 75% = 2026-11-19 — three quarters expire before end of 2026
- Max = 2036 — some very long contracts (consistent with the 120-month duration we saw in orders)
  

#### Key insight for churn definition:

The median `expire_date` being around today means the dataset captures both historical churned subscriptions and currently active ones — it's not just a snapshot of active base, it's the full lifecycle history. 

This is what makes it possible to build proper cohort retention analysis.

```python
today = pd.Timestamp.today()
print((subscriptions['expire_date'] >= today).sum(), "active (expire in future)")
print((subscriptions['expire_date'] < today).sum(), "expired")
```

- 64,881 subscriptions with expire_date in the future (38%) — potential active base
- 105,364 already expired (62%) — historical churn captured in the dataset
- **But `expire_date` alone is not enough to define active** — need to cross with status field, since a subscription can have a future `expire_date` but be explicitly CANCELLED or SUSPENDED
- Next: check how many of the 64,881 "future" subscriptions have a non-null problematic status

```python
future = subscriptions[subscriptions['expire_date'] >= today]
future['status'].value_counts(dropna=False)
```

- True active base is not 64,881 — it's closer to 62,839 (NaN status + future `expire_date`)
- CANCELLED with future `expire_date` = customer cancelled but still within paid period — not active for MRR purposes
- AWAITING PAYMENT + UNPAID = early churn signal, worth tracking separately
- This confirms the active definition needs both conditions: `expire_date >= today AND (status IS NULL OR status NOT IN ('CANCELLED', 'TO DELETE', 'SUSPENDED'))`

### Subscriptions - Actions

- Cast `creation_date`, `start_date`, `expire_date` to datetime in staging
- Derive `subscription_status_clean` field combining `expire_date` and `status` — do not use raw `status` alone
- Active definition: `expire_date >= today AND (status IS NULL OR status NOT IN ('CANCELLED', 'TO DELETE', 'SUSPENDED'))`
- Flag AWAITING PAYMENT + UNPAID separately as early churn signal — useful for a "at risk" segment
- Treat CANCELLED with future `expire_date` as churned for MRR purposes, not active

## Referential integrity

We now check whether the four tables can actually be joined cleanly

Orphan IDs cause loss in the final model without warning

### **`orders` & `products`**

```python
orders_products = set(orders['product_id'].unique())
products_ids = set(products['product_id'].astype(int, errors='ignore'))
print("product_ids in orders not in products:", len(orders_products - products_ids))
```

1462 is the entire products table size, which means the join is failing completely

Root cause: `product_id` in orders is integer but in products it's object (string) — the `astype(int, errors='ignore')` silently failed

```python
products_ids = set(pd.to_numeric(products['product_id'], errors='coerce').dropna().astype(int).unique())
print("product_ids in orders not in products:", len(orders_products - products_ids))
```

The earlier result of 1,462 was a false positive caused by type mismatch (int vs string). 

Real orphan count is 1 (0.0005%). Keep, flag for the team.

### **`orders` & `subscriptions`**

```python
orders_subs = set(orders['subscription_id'].dropna().astype(int).unique())
subs_ids = set(subscriptions['subscription_id'].unique())
print("subscription_ids in orders not in subscriptions:", len(orders_subs - subs_ids))
```

4 subscription_ids in orders not found in subscriptions — 0.002% but worth documenting

These orders will lose their subscription context in the join — keep them, flag for the team

### Referential Integrity - Actions

**`orders` & `products`**
- 1 `product_id` in orders has no match in products — the earlier result of 1,462 was a false positive caused by type mismatch (int vs string). Real orphan count is 1, negligible (0.0005%). Keep, flag for the team.

**`orders` & `subscriptions`**
- 4 subscription_ids in orders not found in subscriptions — 0.002% but worth documenting
- These orders will lose their subscription context in the join — keep them, flag for the team
